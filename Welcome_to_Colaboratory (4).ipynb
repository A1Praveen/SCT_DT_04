```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
file_path = '/content/dt.csv'  # Replace with the correct path to your file
data = pd.read_csv(file_path)

# Display the first few rows of the dataset to understand its structure
print(data.head())

# Basic information about the dataset
print(data.info())

# Summary statistics
print(data.describe())

# Check for missing values
print(data.isnull().sum())

numeric_data=data.select_dtypes(include=['float','int'])


# Correlation matrix to see relationships between numerical variables
corr_matrix = numerical_data.corr()
print(corr_matrix)

# Plotting the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()



# Check if the columns exist and have expected values
if 'Day_of_Week' in data.columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=data, x='Day_of_Week', order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
    plt.title('Number of Accidents by Day of the Week')
    plt.show()
else:
    print("Column 'Day_of_Week' not found in the dataset.")

if 'Weather' in data.columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=data, x='Weather')
    plt.title('Number of Accidents by Weather Condition')
    plt.xticks(rotation=45)
    plt.show()
else:
    print("Column 'Weather_Condition' not found in the dataset.")

# Analyze accidents by speed
if 'Speed' in data.columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(data['Speed'], bins=20, kde=True)
    plt.title('Distribution of Speed in Accidents')
    plt.show()
else:
    print("Column 'Speed' not found in the dataset.")

# Analyze accidents by economic conditions (assuming you have a column for it)
if 'Economic_Condition' in data.columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=data, x='Economic_Condition',y='Accidents')
    plt.title('Number of Accidents by Economic Condition')
    plt.xticks(rotation=45)
    plt.show()
else:
    print("Column 'Economic_Condition' not found in the dataset.")

# Analyze relationship between speed and weather condition
if 'Weather_Condition' in data.columns and 'Speed' in data.columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=data, x='Weather_Condition', y='Speed')
    plt.title('Speed Distribution by Weather Condition')
    plt.xticks(rotation=45)
    plt.show()
else:
    print("Columns 'Weather_Condition' or 'Speed' not found in the dataset.")

# Summarize the data based on traffic density
traffic_density_summary = data.groupby('Traffic Density').agg({
    'City': 'count',
    'Vehicle Type': lambda x: x.value_counts().index[0],  # Most common vehicle type
    'Weather': lambda x: x.value_counts().index[0],       # Most common weather condition
    'Economic Condition': lambda x: x.value_counts().index[0],  # Most common economic condition
    'Day Of Week': lambda x: x.value_counts().index[0],   # Most common day of the week
    'Hour Of Day': lambda x: x.mode()[0],                 # Most common hour of the day
    'Speed': 'mean',                                      # Average speed
    'Is Peak Hour': 'mean',                               # Proportion of peak hour accidents
    'Random Event Occurred': 'mean',                      # Proportion of accidents with random events
    'Energy Consumption': 'mean'                          # Average energy consumption
}).rename(columns={'City': 'Number of Accidents'})

# Display the summary
print(traffic_density_summary)

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data from the provided CSV file
file_path = '/content/dt.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset to understand its structure
print(data.head())

# Summary statistics of the dataset
summary_stats = data.describe(include='all')
print(summary_stats)

# Count plot of accidents by economic condition
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Economic Condition', palette='viridis')
plt.title('Count of Accidents by Economic Condition')
plt.xlabel('Economic Condition')
plt.ylabel('Count')
plt.show()

# Box plot of Speed by Economic Condition
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Economic Condition', y='Speed', palette='viridis')
plt.title('Speed Distribution by Economic Condition')
plt.xlabel('Economic Condition')
plt.ylabel('Speed')
plt.show()

# Count plot of Random Event Occurred by Economic Condition
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Economic Condition', hue='Random Event Occurred', palette='viridis')
plt.title('Random Events by Economic Condition')
plt.xlabel('Economic Condition')
plt.ylabel('Count')
plt.show()

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = '/content/Road accident.csv'
accident_data = pd.read_csv(file_path)

# Clean column names if necessary
accident_data.columns = [
    'collision_index', 'collision_year', 'collision_reference', 'vehicle_reference',
    'casualty_reference', 'casualty_class', 'sex_of_casualty', 'age_of_casualty',
    'age_band_of_casualty', 'casualty_severity', 'pedestrian_location',
    'pedestrian_movement', 'car_passenger', 'bus_or_coach_passenger',
    'pedestrian_road_maintenance_worker', 'casualty_type', 'casualty_home_area_type',
    'casualty_imd_decile', 'lsoa_of_casualty'
]

# Display a summary of the dataset
print(accident_data.describe(include='all'))

# Check for missing values
print(accident_data.isnull().sum())

# If needed, fill or drop missing values
# accident_data = accident_data.dropna()

# Analyze the distribution of accidents by severity
severity_distribution = accident_data['casualty_severity'].value_counts()
print(severity_distribution)

# Plot accidents by severity
plt.figure(figsize=(10, 6))
sns.countplot(x='casualty_severity', data=accident_data, palette='viridis')
plt.xlabel('Casualty Severity')
plt.ylabel('Number of Accidents')
plt.title('Distribution of Accidents by Severity')
plt.show()

# Analyze accident hotspots using LSOA
hotspots = accident_data['lsoa_of_casualty'].value_counts().head(20)
print(hotspots)

# Plot accident hotspots
plt.figure(figsize=(10, 6))
sns.barplot(x=hotspots.values, y=hotspots.index, palette='viridis')
plt.xlabel('Number of Accidents')
plt.ylabel('LSOA of Casualty')
plt.title('Top 20 Accident Hotspots')
plt.show()
```
